```python
from sklearn.metrics.pairwise import cosine_similarity
import sklearn
import json
import numpy
```

```python
inp_file_path = r'general-1.json'
inp_file = open(inp_file_path, 'r')
inp_text = inp_file.read()
qa = json.loads(inp_text)
qa[0:2]
questions = [item['question'] for item in qa]
answers = [item['answer'] for item in qa]
print(questions)
```

```python
def getWordVec(word,model):
        samp=model.wv.get_vector('தந்தை')
        vec=[0]*len(samp)
        try:
                vec=model.wv.get_vector(word)
        except:
                vec=[0]*len(samp)
        return (vec)


def getPhraseEmbedding(phrase,embeddingmodel):            
        samp=getWordVec('தந்தை', embeddingmodel)
        vec=numpy.array([0]*len(samp))
        print("initial", vec)
        for word in phrase:
            vec=vec+numpy.array(getWordVec(word,embeddingmodel))
        return vec.reshape(1, -1)
```

```python
question_embeddings = []
for question in questions:
  question_embeddings.append(getPhraseEmbedding(tokenize(question), model))
```

```python
user_question = "இடுப்பு வலி போக"
user_question_embedding = getPhraseEmbedding(tokenize(user_question), model)
```

```python

def retrieveAndPrintFAQAnswer(user_question_embedding,question_embeddings, questions, answers):
    max_sim=-1
    index_sim=-1
    for index,question_embedding in enumerate(question_embeddings):
        sim=cosine_similarity(question_embedding,user_question_embedding)[0][0]
        print(index, sim, questions[index])
        if sim>max_sim:
            max_sim=sim
            index_sim=index
       
    print("\n")
    print("Question: ",user_question)
    print("\n");
    print("Retrieved: ",questions[index_sim])
    print(answers[index_sim])        
    
retrieveAndPrintFAQAnswer(user_question_embedding,question_embeddings,questions, answers)
```
